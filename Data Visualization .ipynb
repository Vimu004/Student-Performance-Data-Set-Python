```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
ds = pd.read_csv(r"C:\Users\user\Downloads\GroupProject 8(DS in python)\Dataset\(New Dataset)student-por.csv")
ds
```

```python
ds['G1 + G2'] = (ds['G1'] + ds['G2']) / 2 #Create a new attribute using G1 and G2
ds.head()
```

```python
unique_ids = pd.Series(range(1, len(ds) + 1)) #Generate unique IDs for each record
ds['ID'] = unique_ids #Add the unique IDs as a new column to the Dataset
ds.head()
```

```python
ds["G3"].hist(bins=20)
plt.title('G3 - Number of students')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 6))
plt.tight_layout(pad=4)
chosen_columns = ['traveltime','studytime','freetime'] 
axes = axes.flatten()

for i, column in enumerate(chosen_columns):
    axes[i].hist(ds[column], bins=10, color='lightblue', edgecolor='black')
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Histogram of {column}')

```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 5))
plt.tight_layout(pad=4)
chosen_columns = ['G1','G2','G3'] 
axes = axes.flatten()

for i, column in enumerate(chosen_columns):
    axes[i].hist(ds[column], bins=20, color='lightblue', edgecolor='black')
    axes[i].set_xlabel('Grade')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Histogram of {column}')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
plt.tight_layout(pad=4)
chosen_columns = ['Medu','Fedu'] 
axes = axes.flatten()

for i, column in enumerate(chosen_columns):
    axes[i].hist(ds[column], bins=20, color='lightblue', edgecolor='black')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Histogram of {column}')
```

```python
chosen_columns = ['studytime','freetime'] 
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes = axes.flatten()

for i, column in enumerate(chosen_columns):
    unique_values = ds[column].value_counts().index
    counts = ds[column].value_counts().values
    axes[i].bar(unique_values, counts, color='skyblue', edgecolor='black')
    axes[i].set_xlabel('Values')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Bar Plot of {column}')
plt.tight_layout()
```
