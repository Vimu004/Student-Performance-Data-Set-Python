```python
school_counts = dset['school'].value_counts()
plt.figure(figsize=(8, 6))

school_counts.plot(kind='bar', color='skyblue')
plt.title('Student Count in Each School')
plt.xlabel('School')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
school_gp = dset[dset['school'] == 'GP']
school_ms = dset[dset['school'] == 'MS']

# Count number of students in each school based on final grade (G3)
gp_g3_counts = school_gp['G1'].value_counts().sort_index()
ms_g3_counts = school_ms['G2'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(10, 6))

# Plot GP school
plt.bar(gp_g3_counts.index, gp_g3_counts.values, width=0.4, align='center', label='GP', color='skyblue')

# Plot MS school
plt.bar(ms_g3_counts.index + 0.4, ms_g3_counts.values, width=0.4, align='center', label='MS', color='salmon')

plt.title('First Grade(G1) Student Count in Each School')
plt.xlabel('First Grade (G1)')
plt.ylabel('Number of Students')
plt.xticks(range(0, 21))
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
school_freetime_counts = dset.groupby(['school', 'freetime']).size().unstack(fill_value=0)
school_studytime_counts = dset.groupby(['school', 'studytime']).size().unstack(fill_value=0)

# Plotting Free Time and Study Time in Each School
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot Free Time in Each School
school_freetime_counts.plot(kind='bar', stacked=True, ax=axes[0, 0])
axes[0, 0].set_title('Student Count by Free Time in Each School')
axes[0, 0].set_xlabel('School')
axes[0, 0].set_ylabel('Number of Students')
axes[0, 0].legend(title='Free Time', bbox_to_anchor=(1, 1))
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.7)

# Plot Study Time in Each School
school_studytime_counts.plot(kind='bar', stacked=True, ax=axes[0, 1])
axes[0, 1].set_title('Student Count by Study Time in Each School')
axes[0, 1].set_xlabel('School')
axes[0, 1].set_ylabel('Number of Students')
axes[0, 1].legend(title='Study Time', bbox_to_anchor=(1, 1))
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Hide the unused subplot
axes[1, 0].axis('off')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
mother_higher_edu = dset.groupby('Mjob')['higher'].value_counts(normalize=True).unstack()
father_higher_edu = dset.groupby('Fjob')['higher'].value_counts(normalize=True).unstack()

# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Mother's job
mother_higher_edu.plot(kind='bar', stacked=True, ax=ax1)
ax1.set_title("Effect of Mother's Job on Higher Education")
ax1.set_xlabel("Mother's Job")
ax1.set_ylabel('Percentage')
ax1.legend(title='Wants to Take Higher Education', bbox_to_anchor=(1, 1))

# Father's job
father_higher_edu.plot(kind='bar', stacked=True, ax=ax2)
ax2.set_title("Effect of Father's Job on Higher Education")
ax2.set_xlabel("Father's Job")
ax2.set_ylabel('Percentage')
ax2.legend(title='Wants to Take Higher Education', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

```python
dset['Dalc_bin'] = pd.cut(dset['Dalc'], bins=5, labels=['Very Low', 'Low', 'Moderate', 'High', 'Very High'])
dset['Walc_bin'] = pd.cut(dset['Walc'], bins=5, labels=['Very Low', 'Low', 'Moderate', 'High', 'Very High'])

# Group data by daily alcohol consumption and calculate average health status
dalc_health = dset.groupby('Dalc_bin')['health'].mean()

# Group data by weekly alcohol consumption and calculate average health status
walc_health = dset.groupby('Walc_bin')['health'].mean()

# Plotting
plt.figure(figsize=(12, 6))

# Bar graph for Daily Alcohol Consumption vs Health
plt.subplot(1, 2, 1)
dalc_health.plot(kind='bar', color='skyblue')
plt.title('Daily Alcohol Consumption vs Health')
plt.xlabel('Daily Alcohol Consumption')
plt.ylabel('Average Health Status')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Bar graph for Weekly Alcohol Consumption vs Health
plt.subplot(1, 2, 2)
walc_health.plot(kind='bar', color='lightgreen')
plt.title('Weekly Alcohol Consumption vs Health')
plt.xlabel('Weekly Alcohol Consumption')
plt.ylabel('Average Health Status')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python

# Define the order of facets
job_order = ['at_home', 'health', 'other', 'services', 'teacher']
edu_order = [0, 1, 2, 3, 4]

# Create a FacetGrid
g = sns.FacetGrid(dset, row='Medu', col='Mjob', row_order=edu_order, col_order=job_order, height=4, aspect=1)

# Map the histogram distribution of final grade to each facet
g.map(sns.histplot, 'G3', bins=10)

# Set titles and labels
g.set_axis_labels('Final Grade (G3)', 'Frequency')
g.set_titles('Job: {col_name}, Education: {row_name}')
g.set_ylabels('Student count')
g.set_xlabels('Final Grade(G3)')
plt.suptitle('Histogram Distribution of Final Grade Faceted by Mother\'s Job vs. Mother\'s Education', y=1.02)
plt.tight_layout()
plt.show()
```
