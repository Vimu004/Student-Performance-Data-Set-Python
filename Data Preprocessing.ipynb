# IMPORTANT
in  Portugal, During the school year, students are evaluated in three periods(G1, G2 ,G3) and the last evaluation (G3) corresponds to the final grade.

There is an important relationship between attributes G2 and G1 and the target attribute G3. This happens because first period gradeÂ G1 and second periodÂ G2 directly influenceÂ the grades for theÂ G3 is the final year grade (given during the third period). Without G2 and G1, predicting G3 is more challenging.
    
       

```python
sns.histplot(data=ds, x='G1', bins=20, kde=True, color='blue', alpha=0.7)
plt.title('Distribution of First Period Grade')
plt.xlabel('First Period Grade')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=ds['G1'])
plt.title('Box Plot of G1'.format('G1'))
plt.xlabel('G1')
plt.show()
```

```python
sns.histplot(data=ds, x='G2', bins=20, kde=True, color='blue', alpha=0.7)
plt.title('Distribution of Second Period Grade')
plt.xlabel('Second Period Grade')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=ds['G2'])
plt.title('Box Plot of G2'.format('G2'))
plt.xlabel('G2')
plt.show()
```

```python
sns.histplot(data=ds, x='G3', bins=20, kde=True, color='blue', alpha=0.7)
plt.title('Distribution of Third Period Grade/Final Grade')
plt.xlabel('Third Period Grade')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=ds['G3'])
plt.title('Box Plot of G3'.format('G3'))
plt.xlabel('G3')
plt.show()
```
